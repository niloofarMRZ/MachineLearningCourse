```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from fcmeans import FCM
from seaborn import scatterplot as scatter

def make_points(n,off):
    point = np.random.randn(n, 2)
    point_with_offset = point + [0,off]
    return 5*point_with_offset[:, 0], point_with_offset[:, 1]

a,b = make_points(1000,2)
plt.scatter(a,b,c='green')
c,d = make_points(1000,-2)
plt.scatter(c,d,c='red')
plt.yticks(range(-10,11,5))
plt.show()

DATA = []
for i in range(len(a)):
    DATA.append([a[i],b[i]])
    DATA.append([c[i], d[i]])

df = pd.DataFrame(DATA)

plt.scatter(df.iloc[:,0],df.iloc[:,1],c='purple')
plt.yticks(range(-10,11,5))
plt.show()

fcm = FCM(n_clusters=2)
fcm.fit(df)
fcm_centers = fcm.centers
fcm_labels  = fcm.u.argmax(axis=1)

scatter(df.iloc[:,0], df.iloc[:,1],  hue=fcm_labels)
scatter(fcm_centers.iloc[:,0], fcm_centers.iloc[:,1],marker="X",s=200)
plt.yticks(range(-10,11,5))
plt.show()
```
